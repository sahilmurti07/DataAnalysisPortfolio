```python
import matplotlib.pyplot as plt
import seaborn as sns

# Convert 'order_timestamp' to datetime objects
summary['order_timestamp'] = pd.to_datetime(summary['order_timestamp'])

# Extract date from 'order_timestamp'
summary['order_date'] = summary['order_timestamp'].dt.date

# Group by date and count orders
daily_orders = summary.groupby('order_date')['order_id'].nunique().reset_index()
daily_orders.columns = ['order_date', 'num_orders']

# Plot the growth of orders over time
plt.figure(figsize=(12, 6))
sns.lineplot(x='order_date', y='num_orders', data=daily_orders)
plt.title('Daily Number of Orders Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Orders')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
